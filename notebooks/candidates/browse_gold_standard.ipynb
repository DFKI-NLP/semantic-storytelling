# Let's look at the gold standard in a convenient way

```python
import pandas as pd
import os
import json
from IPython.core.display import display, HTML
```

```python
data_dir = '/data/experiments/plakidis/storytelling-candidates/data'

input_fp = os.path.join(data_dir, 'results_18-02-21.json')
```

```python
docs = json.load(open(input_fp))
```

```python
doc_id2doc = {d['doc']: d for d in docs}
```

```python
cols = ['doc', 'title', 'total_n_sources', 'n_uni_sources', 'unique_score', 'plagiarized_percentage']

df = pd.DataFrame([[d[c] for c in cols] for d in docs], columns=cols)
df
```

```python
#sort_by = 'total_n_sources'
sort_by = 'unique_score'
selector = df.n_uni_sources > 5
top_k = 20

print(f'Displaying the top {top_k} of {len(df[selector])} articles')

for doc_id in df[selector].sort_values(sort_by, ascending=False)[:top_k]['doc']:
    doc = doc_id2doc[doc_id]
    
    # view doc
    sent_html = ''

    for sent in doc['sentences']:
        sent = sent['sentence']
        sent_html += f'''
            <b>{sent['sent_text']} ({sent['start']}-{sent['end']})</b>
            <ul>
        '''

        for s in sent['sources']:
            s = s['source']
            sent_html += f'''<li>{s['source_sent_text']} ({s['source_start']}-{s['source_end']}; 
            from <a href="{s['source_url']}" target="_blank">{s['source_title']}</a>)</li>'''

        sent_html += '</ul>'

    html = f'''
    <div>
    <h1><a href="{doc['url']}">{doc['title']}</a></h1>

    <p>Article text: {' '.join(doc['text'])[:250]} ...</p>

    <p><u>Plagiarised sentences:</u></p>

    {sent_html}

    </div>
    <hr />
    '''

    display(HTML(html))
    
    
    
```
